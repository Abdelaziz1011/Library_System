```python
class Library:
    def __init__(self, name):
        
        self.name = name
        self.books_list = []
        self.lent_books = {}

    def display_available_books(self):
        
        available_books = [book for book in self.books_list if book not in self.lent_books.keys()]
        print("Available Books:")
        for book in available_books:
            print(book)

    def lend_book(self, book, user):
        
        if book in self.books_list and book not in self.lent_books.keys():
            self.lent_books[book] = user
            print(f"Book '{book}' has been lent to {user}.")
        else:
            print(f"Sorry, book '{book}' is not available or already lent.")

    def add_book(self, book):
        
        self.books_list.append(book)
        print(f"Book '{book}' has been added to the library.")

    def return_book(self, book, user):
        
        if book in self.lent_books.keys() and self.lent_books[book] == user:
            del self.lent_books[book]
            print(f"Book '{book}' has been returned by {user}.")
        else:
            print(f"Sorry, book '{book}' is not lent by {user} or not found.")
```

```python
library = Library("My Library")

library.add_book("Book A")
library.add_book("Book B")
library.add_book("Book C")

library.display_available_books()

library.lend_book("Book A", "ZEZO")
library.lend_book("Book B", "DIDI")

library.display_available_books()

library.return_book("Book A", "ZEZO")

library.display_available_books()
```
